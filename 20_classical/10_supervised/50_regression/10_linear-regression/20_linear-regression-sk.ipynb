# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.isotonic import IsotonicRegression
from sklearn.utils import check_random_state
```

## Load Data

```python
n = 100
X = np.arange(n)
randomState = check_random_state(0)
y = randomState.randint(-50, 50, size=(n,)) + 50. * np.log(1 + np.arange(n))
print(X.shape)
print(y.shape)
```

## Train

```python
isotonicRegressor = IsotonicRegression()  # 水平回归
y_isotonic = isotonicRegressor.fit_transform(X, y)

linearRegressor = LinearRegression()  # 线性回归
linearRegressor.fit(X[:, np.newaxis], y)  # 线性回归的数据集需要为2d
```

## Figure

```python
fig = plt.figure()
plt.plot(X, y, 'r.', markersize=12)
plt.plot(X, y_isotonic, 'g.-', markersize=12)
plt.plot(X, linearRegressor.predict(X[:, np.newaxis]), 'b-')
plt.legend(('Data', 'Isotonic Fit', 'Linear Fit'), loc='lower right')
plt.title('Isotonic & Linear Regression')
plt.show()
```
