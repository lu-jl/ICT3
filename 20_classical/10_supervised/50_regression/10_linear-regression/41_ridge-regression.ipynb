```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
```

```python
data = pd.read_csv('dataset/ccpp.csv')
```

```python
X = data[['AT', 'V', 'AP', 'RH']]
y = data[['PE']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

è¦è¿è¡Œ Ridge å›å½’ï¼Œå¿…é¡»è¦æŒ‡å®šè¶…å‚æ•° ğ›¼ã€‚ä½ ä¹Ÿè®¸ä¼šé—®ï¼šâ€œæˆ‘ä¹Ÿä¸çŸ¥é“è¶…å‚æ•°æ˜¯å¤šå°‘å•Šï¼Ÿâ€ æˆ‘ä¹Ÿä¸çŸ¥é“ï¼Œé‚£ä¹ˆæˆ‘ä»¬éšæœºæŒ‡å®šä¸€ä¸ª(æ¯”å¦‚1)ï¼Œåé¢æˆ‘ä»¬ä¼šè®²åˆ°ç”¨äº¤å‰éªŒè¯ä»å¤šä¸ªè¾“å…¥è¶…å‚æ•° ğ›¼ ä¸­å¿«é€Ÿé€‰æ‹©æœ€ä¼˜è¶…å‚æ•°çš„åŠæ³•ã€‚

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)
print(ridge.score(X_train, y_train))
```

ä¹Ÿå°±æ˜¯è¯´æˆ‘ä»¬å¾—åˆ°çš„æ¨¡å‹æ˜¯ï¼š
ğ‘ƒğ¸=447.05552892âˆ’1.97373209âˆ—ğ´ğ‘‡âˆ’0.2323016âˆ—ğ‘‰+0.06935852âˆ—ğ´ğ‘ƒâˆ’0.15806479âˆ—ğ‘…ğ»

```python
print(ridge.coef_)
print(ridge.intercept_)
y_pred = ridge.predict(X_test)
from sklearn import metrics
print("MSE:",metrics.mean_squared_error(y_test, y_pred))  # ç”¨scikit-learnè®¡ç®—MSE
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  # ç”¨scikit-learnè®¡ç®—RMSE
```
